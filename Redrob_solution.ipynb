```python
resumes = [
    {"id": 1, "skills": "Pyhton, MachineLearning, SQL, pandas, numpy, Deep-learning"},
    {"id": 2, "skills": "JavaScrpit, Reacts, Node.JS, MongoDb, REST api, HTML/CSS"},
    {"id": 3, "skills": "Java, Spring Boot, MySql, Microservices, Docker, kubernates"},
    {"id": 4, "skills": "Python, TensorFlow, Keras, NLP, BERT, data-viz, matplotlib"},
    {"id": 5, "skills": "C++, Algoritms, Data Structure, competitive programming, python"},
    {"id": 6, "skills": "javascript, vue.js, python, flask, PostgreSQL, AWS, CI/CD"},
    {"id": 7, "skills": "Python, Sklearn, XGboost, feature engineering, SQL, tableau"},
    {"id": 8, "skills": "Java, Android, Kotlin, Firebase, REST, UI/UX, figma"},
    {"id": 9, "skills": "Reactjs, TypeScrpit, GraphQL, redux, tailwind, nodejs, jest"},
    {"id": 10, "skills": "python, R, statistics, ML, regression, clustering, Power-BI"},
]

jds = [
    {"id": 1, "required": ["Python", "Machine Learning", "Deep Learning", "TensorFlow", "PyTorch", "SQL", "Data Visualization"], "preferred": ["NLP", "BERT", "Feature Engineering", "Statistics"]},
    {"id": 2, "required": ["Java", "Spring Boot", "MySQL", "PostgreSQL", "Microservices", "Docker", "Kubernetes"], "preferred": ["REST API", "CI/CD", "Redis"]},
    {"id": 3, "required": ["JavaScript", "React", "Vue", "TypeScript", "REST API", "HTML/CSS"], "preferred": ["Node.js", "GraphQL", "Redux", "Jest", "AWS"]},
]
```

```python
def normalize_skills(resume):
    skills = []
    for skill in resume["skills"].split(','):
        skill = skill.strip().lower()
        if skill in SKILL_ALIASES:
            skills.append(SKILL_ALIASES[skill])
    return skills

normalized_resumes = []
for resume in resumes:
    normalized_skills = normalize_skills(resume)
    normalized_resumes.append({"id": resume["id"], "skills": normalized_skills})
```

```python
deduplicated_resumes = []
for resume in normalized_resumes:
    unique_skills = set()
    for skill in resume["skills"]:
        unique_skills.add(skill)
    deduplicated_resumes.append({"id": resume["id"], "skills": list(unique_skills)})
```

```python

def _normalize_jd_skill_list(skills_list):
    normalized_skills = set()
    for skill in skills_list:
        skill = skill.strip().lower()
        if skill in SKILL_ALIASES:
            normalized_skills.add(SKILL_ALIASES[skill])
    return list(normalized_skills)


normalized_jds = []
for jd in jds:
    normalized_required = _normalize_jd_skill_list(jd["required"])
    normalized_preferred = _normalize_jd_skill_list(jd["preferred"])
    normalized_jds.append({
        "id": jd["id"],
        "required_skills": normalized_required,
        "preferred_skills": normalized_preferred
    })


all_skills = set()

for resume in deduplicated_resumes:
    all_skills.update(resume["skills"])

for jd in normalized_jds:
    all_skills.update(jd["required_skills"])
    all_skills.update(jd["preferred_skills"])


binary_job_descriptions = []
for job_description in normalized_jds:
    binary_vector = {}
    for skill in all_skills:
        if skill in job_description["required_skills"] or skill in job_description["preferred_skills"]:
            binary_vector[skill] = 1
        else:
            binary_vector[skill] = 0

    binary_job_descriptions.append({"id": job_description["id"], "binary_vector": binary_vector})
```

```python
# To calculate cosine similarity, we need two numerical vectors of the same length.
# 'tf_idf_vector' was undefined, and 'binary_vector' was a dictionary.

# Ensure a consistent order for skills by converting 'all_skills' to a sorted list.
# 'all_skills' was populated in the previous cell 'S5Y4bSZGsnX0'.
ordered_all_skills = sorted(list(all_skills))

# 1. Define 'tf_idf_vector' (e.g., as a binary vector for the first resume).
# We'll use the skills from the first deduplicated resume as an example.
first_resume_skills = deduplicated_resumes[0]["skills"]
tf_idf_vector_list = [1 if skill in first_resume_skills else 0 for skill in ordered_all_skills]

# 2. Convert 'binary_vector' (from the last processed job description in S5Y4bSZGsnX0) into an ordered list.
# 'binary_vector' is available in the kernel state as a dictionary.
binary_vector_list = [binary_vector.get(skill, 0) for skill in ordered_all_skills]

# Calculate dot product:
dot_product = sum(t * b for t, b in zip(tf_idf_vector_list, binary_vector_list))

# Calculate magnitudes (Euclidean norm):
magnitude_tf_idf = math.sqrt(sum(t**2 for t in tf_idf_vector_list))
magnitude_binary = math.sqrt(sum(b**2 for b in binary_vector_list))

# Calculate cosine similarity, handling potential division by zero:
cosine_similarity = 0.0
if magnitude_tf_idf * magnitude_binary != 0:
    cosine_similarity = dot_product / (magnitude_tf_idf * magnitude_binary)
```

```python
# Compute cosine similarity between TF-IDF vectors and binary vectors

# First, create TF-IDF (binary) vectors for resumes
# ordered_all_skills and deduplicated_resumes are available from previous cells.
tf_idf_resumes = []
for resume in deduplicated_resumes:
    tf_idf_vector_dict = {}
    for skill in ordered_all_skills: # Use ordered_all_skills for consistent vector length and ordering
        tf_idf_vector_dict[skill] = 1 if skill in resume["skills"] else 0
    tf_idf_resumes.append({"id": resume["id"], "tf_idf_vector": tf_idf_vector_dict})

similarities = []
for resume in tf_idf_resumes:
    for job_description in binary_job_descriptions:
        tf_idf_vector_dict = resume["tf_idf_vector"]
        binary_vector_dict = job_description["binary_vector"]

        # Ensure consistent ordering for calculations by iterating over ordered_all_skills
        tf_idf_values = [tf_idf_vector_dict.get(skill, 0) for skill in ordered_all_skills]
        binary_values = [binary_vector_dict.get(skill, 0) for skill in ordered_all_skills]

        # Compute dot product
        dot_product = sum(tf_val * bin_val for tf_val, bin_val in zip(tf_idf_values, binary_values))

        # Compute Euclidean norms
        tf_idf_norm = math.sqrt(sum(tf_val ** 2 for tf_val in tf_idf_values))
        binary_norm = math.sqrt(sum(bin_val ** 2 for bin_val in binary_values))

        # Compute cosine similarity
        cosine_similarity = 0.0
        if tf_idf_norm * binary_norm != 0:
            cosine_similarity = dot_product / (tf_idf_norm * binary_norm)

        similarities.append({"resume_id": resume["id"], "job_description_id": job_description["id"], "cosine_similarity": cosine_similarity})
```

```python
ranked_resumes = []
for job_description in binary_job_descriptions:
    job_description_id = job_description["id"]
    similarities_for_job_description = [similarity for similarity in similarities if similarity["job_description_id"] == job_description_id]
    ranked_resumes_for_job_description = sorted(similarities_for_job_description, key=lambda x: x["cosine_similarity"], reverse=True)
    ranked_resumes.append({"job_description_id": job_description_id, "ranked_resumes": ranked_resumes_for_job_description})
```

```python
# Output top 3 matching resumes for each job description
for ranked_resumes_for_job_description in ranked_resumes:
    job_description_id = ranked_resumes_for_job_description["job_description_id"]
    top_3_resumes = ranked_resumes_for_job_description["ranked_resumes"][:3]
    print(f"Job Description {job_description_id}:")
    for resume in top_3_resumes:
        print(f"Resume {resume['resume_id']}: {resume['cosine_similarity']}")
    print()
```

```python
# Import necessary libraries
import math

# Define the job descriptions
job_descriptions = [
    {"id": 1, "name": "Kakao (ML Engineer)", "required_skills": ["python", "machine learning", "deep learning"], "preferred_skills": ["tensorflow", "pytorch"]},
    {"id": 2, "name": "Naver (Backend Engineer)", "required_skills": ["java", "spring boot", "mysql"], "preferred_skills": ["postgresql", "docker"]},
    {"id": 3, "name": "Line (Frontend Engineer)", "required_skills": ["javascript", "react", "html/css"], "preferred_skills": ["nodejs", "graphql"]}
]

# Define the resumes
resumes = [
    {"id": 1, "skills": ["python", "machine learning", "deep learning", "tensorflow"]},
    {"id": 2, "skills": ["java", "spring boot", "mysql", "postgresql"]},
    {"id": 3, "skills": ["javascript", "react", "html/css", "nodejs"]},
    {"id": 4, "skills": ["python", "machine learning", "deep learning", "pytorch"]},
    {"id": 5, "skills": ["java", "spring boot", "mysql", "docker"]},
    {"id": 6, "skills": ["javascript", "react", "html/css", "graphql"]},
    {"id": 7, "skills": ["python", "machine learning", "deep learning", "tensorflow"]},
    {"id": 8, "skills": ["java", "spring boot", "mysql", "postgresql"]},
    {"id": 9, "skills": ["javascript", "react", "html/css", "nodejs"]},
    {"id": 10, "skills": ["python", "machine learning", "deep learning", "pytorch"]}
]

# Define the skill aliases
skill_aliases = {
    "python": "python",
    "java": "java",
    "javascript": "javascript",
    "machine learning": "machine learning",
    "deep learning": "deep learning",
    "tensorflow": "tensorflow",
    "pytorch": "pytorch",
    "spring boot": "spring boot",
    "mysql": "mysql",
    "postgresql": "postgresql",
    "react": "react",
    "html/css": "html/css",
    "nodejs": "nodejs",
    "graphql": "graphql",
    "docker": "docker"
}

# Normalize the skills
normalized_resumes = []
for resume in resumes:
    normalized_skills = []
    for skill in resume["skills"]:
        if skill in skill_aliases:
            normalized_skills.append(skill_aliases[skill])
    normalized_resumes.append({"id": resume["id"], "skills": normalized_skills})

# Deduplicate the skills
deduplicated_resumes = []
for resume in normalized_resumes:
    unique_skills = set()
    for skill in resume["skills"]:
        unique_skills.add(skill)
    deduplicated_resumes.append({"id": resume["id"], "skills": list(unique_skills)})

# Construct the vocabulary
all_skills = set()
for resume in deduplicated_resumes:
    for skill in resume["skills"]:
        all_skills.add(skill)

# Construct the TF-IDF vectors
tf_idf_resumes = []
for resume in deduplicated_resumes:
    tf_idf_vector = {}
    for skill in all_skills:
        if skill in resume["skills"]:
            tf_idf_vector[skill] = 1
        else:
            tf_idf_vector[skill] = 0
    tf_idf_resumes.append({"id": resume["id"], "tf_idf_vector": tf_idf_vector})

# Construct the binary vectors for job descriptions
binary_job_descriptions = []
for job_description in job_descriptions:
    binary_vector = {}
    for skill in all_skills:
        if skill in job_description["required_skills"] or skill in job_description["preferred_skills"]:
            binary_vector[skill] = 1
        else:
            binary_vector[skill] = 0
    binary_job_descriptions.append({"id": job_description["id"], "binary_vector": binary_vector})

# Compute the cosine similarity
similarities = []
for resume in tf_idf_resumes:
    for job_description in binary_job_descriptions:
        tf_idf_vector = resume["tf_idf_vector"]
        binary_vector = job_description["binary_vector"]

        dot_product = sum(tf_idf_vector[skill] * binary_vector[skill] for skill in tf_idf_vector)
        tf_idf_norm = math.sqrt(sum(tf_idf_vector[skill] ** 2 for skill in tf_idf_vector))
        binary_norm = math.sqrt(sum(binary_vector[skill] ** 2 for skill in binary_vector))

        cosine_similarity = dot_product / (tf_idf_norm * binary_norm)

        similarities.append({"resume_id": resume["id"], "job_description_id": job_description["id"], "cosine_similarity": cosine_similarity})

# Rank the resumes
ranked_resumes = []
for job_description in binary_job_descriptions:
    job_description_id = job_description["id"]
    similarities_for_job_description = [similarity for similarity in similarities if similarity["job_description_id"] == job_description_id]
    ranked_resumes_for_job_description = sorted(similarities_for_job_description, key=lambda x: x["cosine_similarity"], reverse=True)
    ranked_resumes.append({"job_description_id": job_description_id, "ranked_resumes": ranked_resumes_for_job_description})

# Output the final results
for ranked_resumes_for_job_description in ranked_resumes:
    job_description_id = ranked_resumes_for_job_description["job_description_id"]
    job_description_name = job_descriptions[job_description_id - 1]["name"]
    top_3_resumes = ranked_resumes_for_job_description["ranked_resumes"][:3]

    print(f"JD-{job_description_id} — {job_description_name}")
    for i, resume in enumerate(top_3_resumes):
        if i < 2:
            print(f"{resume['resume_id']}({resume['cosine_similarity']:.2f})", end=", ")
        else:
            print(f"{resume['resume_id']}({resume['cosine_similarity']:.2f})")
    print()
```
